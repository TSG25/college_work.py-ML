```python
import pandas as pd
import os
```

```python
df = pd.read_csv("Pandas and numpy task/SalesAnalysis/Sales_Data/Sales_April_2019.csv")


files = [file for file in os.listdir("Pandas and numpy task/SalesAnalysis/Sales_Data")]

all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("Pandas and numpy task/SalesAnalysis/Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data,df])

all_months_data.to_csv("All_data.csv",index=False)

```

```python
total_data = pd.read_csv("All_data.csv")
total_data.head()
```

```python
total_data
```

```python
total_data['month'] = total_data["Order Date"].str[0:2]

total_data.head()
```

```python
total_data.dropna()
```

```python
nan_df = total_data[total_data.isna().any(axis=1)]
nan_df
```

```python
total_data = total_data.dropna(how='any')
```

```python
total_data
```

```python
total_data['Quantity Ordered'] = pd.to_numeric(total_data['Quantity Ordered'])
total_data['Price Each']=pd.to_numeric(total_data['Price Each'])
total_data.head()
```

```python
total_data = total_data[total_data['Order Date'].str[0:2] !='Or']
total_data.head()
```

```python
total_data['month']= total_data['month'].astype('int32')
total_data
```

```python
#task 3 add sales column

total_data['Sales'] = total_data['Quantity Ordered']*total_data['Price Each']
```

```python
total_data.head()
```

```python
#what was the best month for sales? how much earn that month

result = total_data.groupby("month").sum()
```

```python
import matplotlib .pyplot as plt
months = range(1,13)


plt.bar(months ,result['Sales'],color = 'r')
plt.xticks(months)
plt.ylabel("Sales of us dollers")
plt.xlabel("months number")
plt.yticks()
plt.show()
```

```python


#create a nuw colum and use apply method
def get_city(addres):
    return addres.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

total_data['city']= total_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
total_data.head()

```

```python
total_data.loc[:, "city"].unique()
```

```python
#what city had the highest number of sales?

result = total_data.groupby("city").sum()
result.head()
```

```python
result.loc[: ,'Sales']
```

```python
cities = [city for city ,df in total_data.groupby("city")]

plt.bar(cities ,result['Sales'],color = 'g')
plt.xticks(cities ,rotation='vertical')
plt.ylabel("Sales of us dollers")
plt.xlabel(" cites names")
plt.yticks( )
plt.show()
```

```python
total_data['Hour']=total_data['Order Date'].dt.hour
total_data['Minute']=total_data['Order Date'].dt.minute
total_data.head()
```
