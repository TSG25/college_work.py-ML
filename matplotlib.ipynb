```python
import matplotlib .pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5),dpi=200)
plt.title( 'Gas prices over the time',fontdict={'fontsize':20})
plt.plot(gas.Year,gas.USA ,'r.-',label = 'Usa')
plt.plot(gas.Year,gas.Canada ,'b.-',label = 'Canada')
plt.plot(gas.Year,gas.Japan ,'g.-',label='Japan')
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel("Years")
plt.ylabel("Us Dollers $")
plt.legend()
plt.savefig("firstfig.png",dpi=100)
plt.show()
```

```python
plt.figure(figsize=(8,5),dpi=300)
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker=".",label= country)

plt.legend()
plt.show()
```

```python

x_v = np.array([1,3,4,6,7,9,10,12])
y_v =x_v **2
z_v = x_v **3
plt.title('Graph' , fontdict={'color':'red','size': 20})
plt.plot(x_v,y_v,color = "g",marker="o",linestyle = "--",mec = "r",ms=20,label="teja")
plt.plot(x_v,z_v,color = "b",marker="*",linestyle = "-",mfc="y",ms=15)
plt.xlabel('axis')
plt.ylabel('y_axis')
plt.legend()
plt.show()
```
